```python
!pip install sentence-transformers matplotlib seaborn

from sentence_transformers import SentenceTransformer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sentences = [

    "The bowler took three wickets in one over",
    "The batsman scored a century in the match",
    "The team won the cricket match easily",

    "I cooked pasta with tomato sauce",
    "The recipe requires fresh vegetables",
    "Baking a cake needs patience and precision",

    "Cybersecurity protects systems from attacks",
    "Strong passwords help prevent hacking",
    "Encryption ensures data security",

    "Firewalls block unauthorized access"
]

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)

similarity_matrix = np.inner(embeddings, embeddings)

plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix, annot=True)
plt.title("Cosine Similarity Matrix")
plt.show()

query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

scores = np.inner(query_embedding, embeddings)[0]

top_indices = scores.argsort()[-3:-1][::-1]

print("Top 2 similar sentences:")
for i in top_indices:
    print(sentences[i], "Score:", scores[i])
```
